```python
final_data=pd.read_csv('/Users/raunakpandey/Documents/programming/projects/med-flamingo/ViLT/final_data.csv')

# get the unique answers
unique_answers_df = final_data['ans-hyp'].str.lower().unique()
# get the unique answers
unique_answers = df['Medical Word'].unique()
```

```python
from torch.nn.functional import one_hot

# create a dictionary mapping each answer to its index
answer_to_index = {answer: index for index, answer in enumerate(unique_answers)}

# create a list of one-hot vectors for each answer
one_hot_vectors = [one_hot(torch.tensor(answer_to_index[answer]), len(unique_answers)) for answer in unique_answers]
```

```python
# convert the one-hot vectors back to the original words
predicted_answers = [unique_answers[tensor.argmax().item()] for tensor in one_hot_vectors]
```
